```python
                                            ####problème d'optimisation
    
from docplex.mp.model import Model
import matplotlib.pyplot as plt

lines = input("Entrez les lignes (séparées par des virgules): "
             ).split(",")
years = list(range(1, int(input("Entrez le nombre d'années: ")
                         ) + 1))
bus_types = input("Entrez les types de bus (séparés par des virgules): "
                 ).split(",")
last_year = max(years)

nl_k = {(l, k): int(input(f"Entrez la valeur pour nl_k[{l}, {k}]: ")
                   ) for l in lines for k in bus_types}

B_ONC = {k: float(input(f"Entrez la valeur pour B_ONC[{k}]: "))
         for k in bus_types}
C_ONC = float(input("Entrez la valeur pour C_ONC: "))
B_OPC = {k: float(input(f"Entrez la valeur pour B_OPC[{k}]: "))
         for k in bus_types}
C_OPC = float(input("Entrez la valeur pour C_OPC: "))
B_IMC = {k: float(input(f"Entrez la valeur pour B_IMC[{k}]: "))
         for k in bus_types}
B_Diesel = {k: float(input(f"Entrez la valeur pour B_Diesel[{k}]: "))
            for k in bus_types}

K_ONC_l = {l: float(input(f"Entrez la valeur pour K_ONC_l[{l}]: "))
           for l in lines}
K_OPC_l = {l: float(input(f"Entrez la valeur pour K_OPC_l[{l}]: "))
           for l in lines}
K_IMC_l = {l: float(input(f"Entrez la valeur pour K_IMC_l[{l}]: "))
           for l in lines}
K_Diesel_l = {l: float(input
                (f"Entrez la valeur pour K_Diesel_l[{l}]: ")) for l in lines}

eta = float(input("Entrez la valeur pour eta: "))
p_d = float(input("Entrez la valeur pour p_d: "))
p_e = float(input("Entrez la valeur pour p_e: "))

V_t_k = {(t, k): int(input(f"Entrez la valeur pour V_t_k[{t}, {k}]: ")
                    ) for t in years for k in bus_types}

I_max = {t: float(input(f"Entrez la valeur pour I_max[{t}]: ")
                 ) for t in years}
                             
# Définition du modèle
                             
mdl = Model("Electrification")

ONC = mdl.binary_var_matrix(lines, years, name="ONC")
OPC = mdl.binary_var_matrix(lines, years, name="OPC")
IMC = mdl.binary_var_matrix(lines, years, name="IMC")
Diesel = mdl.binary_var_matrix(lines, years, name="Diesel")

CAPEX = mdl.sum(
    (ONC[l, last_year] * nl_k[l, k] * (B_ONC[k] + C_ONC) +
     OPC[l, last_year] * (nl_k[l, k] * B_OPC[k] + 2 * C_OPC) +
     IMC[l, last_year] * nl_k[l, k] * B_IMC[k]+
     Diesel[l, last_year] * nl_k[l, k] * B_Diesel[k] )
    for l in lines for k in bus_types
)

OPEX = mdl.sum(
    (Diesel[l, t] * (K_Diesel_l[l] / eta) * 365 * p_d +
     ONC[l, t] * K_ONC_l[l] * 365 * p_e +
     OPC[l, t] * K_OPC_l[l] * 365 * p_e + 
     IMC[l, t] * K_IMC_l[l] * 365 * p_e)
    for l in lines for t in years
)

mdl.minimize(CAPEX + OPEX)

for l in lines:
    for t in years:
        mdl.add_constraint(ONC[l, t] + OPC[l, t] + IMC[l, t] + Diesel[l, t] == 1)

        if t < last_year:
            mdl.add_constraint(Diesel[l, t] >= Diesel[l, t + 1])

            for X in [ONC, OPC, IMC]:
                mdl.add_constraint(X[l, t] <= X[l, t + 1])

for X in [ONC, OPC, IMC]:
    for l in lines:
        mdl.add_constraint(X[l, last_year] <= 1)

for t in years:
    for k in bus_types:
        for X in [ONC, OPC, IMC, Diesel]:
            mdl.add_constraint(mdl.sum(X[l, t] * nl_k[l, k] 
                                       for l in lines) <= V_t_k[t, k])

for t in years[1:]:
    mdl.add_constraint(
        mdl.sum(
            (ONC[l, t] - ONC[l, t - 1]) * nl_k[l, k] * (B_ONC[k] + C_ONC) +
            (OPC[l, t] - OPC[l, t - 1]) * (nl_k[l, k] * B_OPC[k] + 2 * C_OPC) +
            (IMC[l, t] - IMC[l, t - 1]) * nl_k[l, k] * B_IMC[k]
            for l in lines
            for k in bus_types
        ) <= I_max[t]
    )
# Ajouter une contrainte pour exclure Diesel comme solution
for l in lines:
    for t in years:
        mdl.add_constraint(Diesel[l, t] == 0)  
    
# Résoudre le modèle
solution = mdl.solve()

# Afficher la solution
if solution:
    print("Solution found:")
    print(solution)
    # Visualisation
    technology_colors = {"ONC": "blue", "OPC": "green", "IMC": "red",
                         "Diesel": "orange"}

    fig, ax = plt.subplots()

    for l in lines:
        transition_year = None
        for t in years:
            for tech, tech_vars in zip(["ONC", "OPC", "IMC", "Diesel"],
                                       [ONC, OPC, IMC, Diesel]):
                value = tech_vars[l, t].solution_value
                if value > 0:
                    if tech != "Diesel" and transition_year is None:
                        transition_year = t
                    color = technology_colors[tech] if t >= transition_year
                                else technology_colors["Diesel"]
                    ax.bar(l, 1, color=color, label=tech)
                    break

    ax.set_xticks(range(len(lines)))
    ax.set_xticklabels(lines)
    ax.set_xlabel("Lines")
    ax.set_ylabel("valeur de décision")
    ax.set_title("la meilleur technologie pour chaque ligne")

    # Supprimer les marques et les étiquettes de l'axe y sauf pour 1
    ax.set_yticks([1])
    ax.set_yticklabels([1])

    # Ajouter une légende
    legend_handles = [plt.Rectangle((0, 0), 1, 1, color=color)
                      for color in technology_colors.values()]
    legend_labels = technology_colors.keys()
    ax.legend(legend_handles, legend_labels, loc="center left",
              bbox_to_anchor=(1, 0.5))

    plt.show()
else:
    print("No solution found.")
```

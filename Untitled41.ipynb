```python
#modèle de consommation de traction
import math
import matplotlib.pyplot as plt

# Parameters
m = float(input("Veuillez entrer la masse du bus (en kg) : "))
g = float(input("Veuillez entrer l'accélération gravitationnelle (en m/s^2) : "))
mu_s = float(input("Veuillez entrer le coefficient de résistance au roulement : "))
alpha = float(input("Veuillez entrer l'inclinaison de la route (en radians) : "))
rho = float(input("Veuillez entrer la densité de l'air (en kg/m^3) : "))
c_f = float(input("Veuillez entrer le coefficient de traînée : "))
A = float(input("Veuillez entrer la surface frontale du bus (en m^2) : "))
a = float(input("Veuillez entrer l'accélération (en m/s^2) : "))
delta_T = int(input("Veuillez entrer l'intervalle de temps (en secondes) : "))
v_moy = float(input("Veuillez entrer la vitesse moyenne (en m/s) : "))
q = float(input("Veuillez entrer le facteur de récupération d'énergie : "))

def calculate_consumption(m, g, mu_s, alpha, rho, c_f, A, a, delta_T, v_moy, q,
                          is_electric):
    delta_t = int((delta_T - math.sqrt(delta_T**2 - 4 * v_moy * delta_T / a)) / 2)
    xi = 1 if math.sin(alpha) >= 0 else 0
    C_cons_cyc = 0
    v_max = a*delta_t
    for t in range(0, delta_T):
        if t < delta_t:
            x_double_dot = a
            x_dot = a * t
            distance = 0.5 * a * t**2
        elif t < delta_T - delta_t:
            x_double_dot = 0
            x_dot = v_max
            distance = v_max*t-a*delta_t**2/2
        else:
            x_double_dot = -a
            x_dot =  a * (delta_T - t)
            distance = (-0.5 * a * t**2  +  a * delta_T * t - 0.5 * a * delta_T**2 +
                        a * delta_T * delta_t - a*delta_t**2)

        W_frot = m * g * mu_s * math.cos(alpha) * distance
        W_aero =  0.5 * rho * c_f * A * (a * distance**2 / 2 + a**2 *delta_t**2 *
                (delta_T * v_moy)+2 * a**2 * (delta_T - delta_t) * distance - a
                * distance * (2 * (delta_T * v_moy) - distance))
        W_grav = m * g * math.sin(alpha) * distance
        
        C_cons_src = m * x_double_dot * distance + W_frot + xi * W_grav + W_aero
        
        if is_electric:
            C_recup = q * ((xi - 1) * W_grav + m * a * distance)
            C_cons_cyc += C_cons_src - C_recup
        else:
            C_cons_cyc += C_cons_src      
    ct = C_cons_cyc /delta_T*v_moy*1e-3/ 3.6e6   # kWh/km  
    return ct


# Calculer la consommation pour les bus diesel et électriques
ct_diesel = calculate_consumption(m, g, mu_s, alpha, rho, c_f, A, a, delta_T
                                  , v_moy, q, False)

ct_electric = calculate_consumption(m, g, mu_s, alpha, rho, c_f, A, a, delta_T,
                                    v_moy, q, True)
print("Consommation diesel :", format(ct_diesel, ".2f"), "kwh/km")
print("Consommation électrique :", format(ct_electric, ".2f"), "kWh/km")

# Tracez la consommation d'énergie

def plot_energy_consumption(ct_electric, ct_diesel):
    x_values = ['Electric Bus', 'Diesel Bus']
    y_values = [ct_electric, ct_diesel]
    colors = ['#1f77b4', '#ff7f0e']

    fig, ax = plt.subplots()
    ax.bar(x_values, y_values, color=colors)

    ax.set_ylabel('Energy Consumption (kWh/km)')
    ax.set_title('Energy Consumption per km')
    ax.set_xlabel('Bus Type')

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.yaxis.grid(True)

    plt.tight_layout()
    plt.show()

plot_energy_consumption(ct_electric, ct_diesel)
```

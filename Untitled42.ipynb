```python
import numpy as np

# Saisie des consommations mensuelles pour chaque cas et des températures
consommations_chauffage = []
consommations_refroidissement = [] 
consommations_auxiliaire = []
temperatures = []

# Saisie des données pour chaque mois de l'année
for mois in range(1, 13):
    consommation_chauffage = float(input(
        f"Saisir la consommation de chauffage pour le mois {mois} : "))
    consommations_chauffage.append(consommation_chauffage)
    
    consommation_refroidissement = float(input(
        f"Saisir la consommation de refroidissement pour le mois {mois} : "))
    consommations_refroidissement.append(consommation_refroidissement)
    
    consommation_auxiliaire = float(input(
        f"Saisir la consommation auxiliaire pour le mois {mois} : "))
    consommations_auxiliaire.append(consommation_auxiliaire)
    
    temperature = float(input(f"Saisir la température pour le mois {mois} : "))
    temperatures.append(temperature)

# Ajustement d'un polynôme du second ordre pour chaque cas
coefficients_chauffage = np.polyfit(temperatures, consommations_chauffage, 2)
coefficients_refroidissement = np.polyfit(
    temperatures, consommations_refroidissement, 2)

coefficients_auxiliaire = np.polyfit(temperatures, consommations_auxiliaire, 2)

# Extraction des coefficients
a_chauffage = coefficients_chauffage[0]
b_chauffage = coefficients_chauffage[1]
c_chauffage = coefficients_chauffage[2]

a_refroidissement = coefficients_refroidissement[0]
b_refroidissement = coefficients_refroidissement[1]
c_refroidissement = coefficients_refroidissement[2]

a_auxiliaire = coefficients_auxiliaire[0]
b_auxiliaire = coefficients_auxiliaire[1]
c_auxiliaire = coefficients_auxiliaire[2]

# Affichage des coefficients
print("Coefficients de consommation pour le chauffage :")
print("Coefficient a:", a_chauffage)
print("Coefficient b:", b_chauffage)
print("Coefficient c:", c_chauffage)
print()

print("Coefficients de consommation pour le refroidissement :")
print("Coefficient a:", a_refroidissement)
print("Coefficient b:", b_refroidissement)
print("Coefficient c:", c_refroidissement)
print()

print("Coefficients de consommation pour les équipements auxiliaires :")
print("Coefficient a:", a_auxiliaire)
print("Coefficient b:", b_auxiliaire)
print("Coefficient c:", c_auxiliaire)
```
